## Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn import ensemble
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures,MinMaxScaler
from sklearn.metrics import confusion_matrix,classification_report,f1_score
from sklearn import metrics
from sklearn.model_selection import learning_curve
```

```python
model1=ensemble.AdaBoostClassifier(random_state=0,n_estimators=150)
model2=ensemble.BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=0),n_estimators=100,random_state=0)
model3=ensemble.ExtraTreesClassifier(min_samples_leaf=1, min_samples_split=2,random_state=0, max_depth=1000,n_estimators=1000)
model4=ensemble.GradientBoostingClassifier(random_state=0)
model5=ensemble.RandomForestClassifier(100,max_depth=100,random_state=0,class_weight='balanced')
model6=XGBClassifier(random_state=0,learning_rate=0.01)
model7=make_pipeline(MinMaxScaler(),PolynomialFeatures(1),LogisticRegression(random_state=0, class_weight='balanced'))
model8=make_pipeline(MinMaxScaler(),PolynomialFeatures(1),SVC(random_state=0,class_weight='balanced'))
models=[model1,model2,model3,model4,model5,model7,model6,model8]
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=0)
print(x_train.shape,x_test.shape)
```

```python
def build_model(x_train,y_train):
    for model in models:
        print('Classifier :',model.__class__.__name__)
        pred=model.fit(x_train,y_train).predict(x_test)
        print(classification_report(y_test,pred))
        print(confusion_matrix(y_test,pred))
        print('--'*60)
```

```python
build_model(x_train,y_train)
```

## --------------------Feature_Selection-----------------------------

```python
from sklearn.feature_selection import SelectKBest,SelectFromModel,RFECV,VarianceThreshold,f_classif,chi2
from sklearn.decomposition import PCA
```

```python
def selectfromModel(features,target):
    x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=0)
    selector=SelectFromModel(DecisionTreeClassifier(random_state=0),threshold='mean')
    selector.fit_transform(x_train,y_train)
    new_features=features[features.columns[selector.get_support()]]
    return new_features
```

```python
new=selectfromModel(features,target)
new.shape
x_train,x_test,y_train,y_test=train_test_split(new,target,test_size=0.2,random_state=0)
print(x_train.shape,x_test.shape)
build_model(x_train,y_train)
```

```python
def Selectkbest(features,target,k):
    selector=SelectKBest(f_classif,k=k)
    selector.fit_transform(features,target)
    new_features=features[features.columns[selector.get_support()]]
    return new_features
```

```python
new=Selectkbest(features,target,112)
x_train,x_test,y_train,y_test=train_test_split(new,target,test_size=0.2,random_state=0)
print(x_train.shape,x_test.shape)
build_model(x_train,y_train)
```

```python
def variancethreshold(features,target):
    x=MinMaxScaler().fit_transform(features)
    selector=VarianceThreshold(threshold=x.var(axis=0).mean()-x.var(axis=0).mean()/1.18)
    selector.fit_transform(x,target)
    new_features=features[features.columns[selector.get_support()]]
    return new_features
```

```python
new=variancethreshold(features,target)
new.shape
x_train,x_test,y_train,y_test=train_test_split(new,target,test_size=0.2,random_state=0)
print(x_train.shape,x_test.shape)
build_model(x_train,y_train)
```

```python
from sklearn.ensemble import VotingClassifier
model1=ensemble.AdaBoostClassifier(random_state=0,n_estimators=150)
model2=ensemble.BaggingClassifier(random_state=0)
model3=ensemble.ExtraTreesClassifier(random_state=0)
model4=ensemble.GradientBoostingClassifier(random_state=0)
model5=ensemble.RandomForestClassifier(random_state=0,class_weight='balanced')
model6=XGBClassifier(random_state=0,learning_rate=0.01)
model7=make_pipeline(StandardScaler(),PolynomialFeatures(1),LogisticRegression(random_state=0, class_weight='balanced'))
model8=make_pipeline(StandardScaler(),PolynomialFeatures(1),SVC(random_state=0,class_weight='balanced'))
model9=VotingClassifier([('Ada',model1),('Bag',model2),('Ext',model3),('Grad',model4),
                        ('RDF',model5),('XGB',model6),('Log',model7),('SVC',model8)],voting='hard')
v=[model1,model2,model3,model4,model5,model7,model6,model8,model9]
for mode in v:
    print('Classifier :',mode.__class__.__name__)
    pred=mode.fit(x_train,y_train).predict(x_test)
    print(classification_report(y_test,pred))
    print(confusion_matrix(y_test,pred))
  
    print('--'*60)
```
